```python
from astropy.io import fits
from astropy.nddata import CCDData
from ccdproc import *
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#View Header of Science Image
image = fits.open('science_1.fits')
header_image = dict(image[0].header)
print(f"Exposure time: {header_image.get('EXPTIME')} seconds")
```

```python
#View Header of Dark Frames and Output the exposure times
for i in range(2604,2624):
    dark_image = fits.open(f'darks.0000{i}.DARK.FIT')
    header_dark = dict(dark_image[0].header)
    print(f"Exposure Time for Dark Frame {i}: {header_dark.get('EXPTIME')}")
```

```python
#Output Dark Frames Temperature
for i in range(2604,2624):
    dark_image = fits.open(f'darks.0000{i}.DARK.FIT')
    header_dark = dict(dark_image[0].header)
    print(f"Temperatures for Dark Frame {i}: {header_dark.get('CCD-TEMP')}")
```

```python
ccd1 = CCDData.read('darks.00002604.DARK.FIT', unit='adu')
ccd2 = CCDData.read('darks.00002605.DARK.FIT', unit='adu')
ccd3 = CCDData.read('darks.00002606.DARK.FIT', unit='adu')
ccd4 = CCDData.read('darks.00002607.DARK.FIT', unit='adu')
ccd5 = CCDData.read('darks.00002608.DARK.FIT', unit='adu')
ccd6 = CCDData.read('darks.00002609.DARK.FIT', unit='adu')
ccd7 = CCDData.read('darks.00002610.DARK.FIT', unit='adu')
ccd8 = CCDData.read('darks.00002611.DARK.FIT', unit='adu')
ccd9 = CCDData.read('darks.00002612.DARK.FIT', unit='adu')
ccd10 = CCDData.read('darks.00002613.DARK.FIT', unit='adu')
ccd11 = CCDData.read('darks.00002614.DARK.FIT', unit='adu')
ccd12 = CCDData.read('darks.00002615.DARK.FIT', unit='adu')
ccd13 = CCDData.read('darks.00002616.DARK.FIT', unit='adu')
ccd14 = CCDData.read('darks.00002617.DARK.FIT', unit='adu')
ccd15 = CCDData.read('darks.00002618.DARK.FIT', unit='adu')
ccd16 = CCDData.read('darks.00002619.DARK.FIT', unit='adu')
ccd17 = CCDData.read('darks.00002620.DARK.FIT', unit='adu')
ccd18 = CCDData.read('darks.00002621.DARK.FIT', unit='adu')
ccd19 = CCDData.read('darks.00002622.DARK.FIT', unit='adu')
ccd20 = CCDData.read('darks.00002623.DARK.FIT', unit='adu')

combiner = Combiner([ccd1, ccd2, ccd3, ccd4, ccd5, ccd6, ccd7, ccd8, ccd9, ccd10, ccd11, ccd12, ccd13, ccd14, ccd15, ccd16, ccd17, ccd18, ccd19, ccd20])
combined_median = np.array(combiner.median_combine())
```

```python
hdu_flats = fits.PrimaryHDU(combined_median)
hdul_flats = fits.HDUList([hdu_flats])
hdu_flats.writeto('master_dark_frame.DARK.FIT')
```

```python
master_dark_image = CCDData.read('master_dark_frame.DARK.FIT', unit='adu')
plt.imshow(master_dark_image.data, cmap='inferno', vmin=96.5, vmax=120)
plt.colorbar()
```

```python
subtraction_images = image[0].data - master_dark_image.data
```

```python
hdu_im = fits.PrimaryHDU(subtraction_images)
hdul_im = fits.HDUList([hdu_im])
hdu_im.writeto('subtraction_frame.FIT')
```

```python
subtraction_frame = CCDData.read('subtraction_frame.FIT', unit='adu')
plt.imshow(subtraction_frame.data, cmap='inferno', vmin=96.5, vmax=120)
plt.colorbar()
```

```python
mean_flats = np.array([])
for i in range(2625,2635):
    flats = np.array(CCDData.read(f'flats.0000{i}.FIT', unit='adu'))
    mean_flats = np.append(mean_flats, np.mean(flats))
```

```python
ccd21 = CCDData.read('flats.00002625.FIT', unit='adu')
ccd22 = CCDData.read('flats.00002626.FIT', unit='adu')
ccd23 = CCDData.read('flats.00002627.FIT', unit='adu')
ccd24 = CCDData.read('flats.00002628.FIT', unit='adu')
ccd25 = CCDData.read('flats.00002629.FIT', unit='adu')
ccd26 = CCDData.read('flats.00002630.FIT', unit='adu')
ccd27 = CCDData.read('flats.00002631.FIT', unit='adu')
ccd28 = CCDData.read('flats.00002632.FIT', unit='adu')
ccd29 = CCDData.read('flats.00002633.FIT', unit='adu')
ccd30 = CCDData.read('flats.00002634.FIT', unit='adu')

combiner_flats = Combiner([ccd21, ccd22, ccd23, ccd24, ccd25, ccd26, ccd27, ccd28, ccd29, ccd30])
combined_median_flat = np.array(combiner_flats.median_combine())
```

```python
hdu_flats = fits.PrimaryHDU(combined_median_flat)
hdul_flats = fits.HDUList([hdu_flats])
hdu_flats.writeto('master_flat_frame.FIT')
```

```python
master_flat_image = CCDData.read('master_flat_frame.FIT', unit='adu')
plt.imshow(master_flat_image.data, cmap='inferno', vmin=30135, vmax=33781)
plt.colorbar()
```

```python
mode_flats = np.array(stats.mode(master_flat_image.data))
normed_mode_flats = master_flat_image.data/mode_flats[0][0]
```

```python
hdu_normed_flats = fits.PrimaryHDU(normed_mode_flats)
hdul_normed_flats = fits.HDUList([hdu_normed_flats])
hdu_normed_flats.writeto('normed_flat_frame.FIT')
```

```python
master_normed_flat_image = CCDData.read('normed_flat_frame.FIT', unit='adu')
mu = np.mean(master_normed_flat_image.data)
std = np.std(master_normed_flat_image.data)
plt.imshow(master_normed_flat_image.data, cmap='inferno', vmin=mu-3*std, vmax=mu+3*std)
plt.colorbar()
```

```python
reduced_frame = subtraction_frame.data/master_normed_flat_image.data
```

```python
hdu_reduced_flats = fits.PrimaryHDU(reduced_frame)
hdul_reduced_flats = fits.HDUList([hdu_reduced_flats])
hdu_reduced_flats.writeto('reduced_flats_frame.FIT')
```

```python
reduced_image = CCDData.read('reduced_flats_frame.FIT', unit='adu')
mu_r = np.mean(reduced_image.data)
std_r = np.std(reduced_image.data)
plt.imshow(reduced_image.data, cmap='inferno', vmin=mu_r-3*std_r, vmax=mu_r+3*std_r)
plt.colorbar()
```
